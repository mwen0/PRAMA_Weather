# l'introduction et statistiques descriptive
#### Le but de notre projet c’est d’utiliser la base de donnée de ‘Rain in Australia’ pour prédire s’il pleuvra à l’avenir.
#### On a téléchargé les données sur le site 'Kaggle': https://www.kaggle.com/jsphyg/weather-dataset-rattle-package
#### Cet ensemble de données contient environ dix ans d’observations métrologiques quotidiennes des stations météorologiques australiennes. 


```python
#importation des librairies
import pandas as pd
import os
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
#importation data 
df=pd.read_csv("/Users/miyawen/Documents/python/PRAMA_weather/weatherAUS.csv")
df.head()
```

```python
#remplacer yes/no par 1/0
df['RainToday'].replace({'No': -1, 'Yes': 1},inplace=True)
df['RainTomorrow'].replace({'No': -1, 'Yes': 1},inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='RainToday', palette = 'Set2')
```

```python
# Correlation between each 2 elements
corr = df.corr()
fig = plt.figure(figsize=(8,6))
sns.heatmap(corr)
```

```python
# Correlation between RainTomorrow and other elements
corr.sort_values(by=["RainTomorrow"],ascending=False).iloc[0].sort_values(ascending=False)
```

```python
plt.figure(figsize=(4,4))
sns.FacetGrid(df, hue="RainTomorrow", size=8).map(sns.kdeplot, "Humidity3pm").add_legend()
#plt.ioff() 
plt.show() 
```
